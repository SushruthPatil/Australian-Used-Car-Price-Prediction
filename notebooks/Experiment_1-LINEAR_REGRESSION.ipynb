## C. Train Machine Learning Model

### C.1 Import Algorithm



```python
from sklearn.linear_model import LinearRegression
```

```python
algorithm_selection_explanations = """

I am using Multivariate Linear Regression for this experiment. It finds the best straight line
through the data by learning a coefficient for each feature that minimises the prediction error.

It is a good starting point as it is a simple and interpretable model.

It will work well as there is a linear relationship with the chosen features and the target variable,
price. Features like kilometers_driven showed a clear directional relationship with price in EDA which
linear regression can capture.

The results from this experiment will also serve as a benchmark for the upcoming experiments.

"""
```

```python
# DO NOT MODIFY THE CODE IN THIS CELL
print_tile(size="h3", key='algorithm_selection_explanations', value=algorithm_selection_explanations)
```

### C.2 Set Hyperparameters


```python
print("Hyperparameters to test:")
print("Model 1: fit_intercept = True")
# model with intercept
model_true = LinearRegression(fit_intercept=True)
print("Model 2: fit_intercept = False")
# model without intercept
model_false = LinearRegression(fit_intercept=False)
```

```python
hyperparameters_selection_explanations = """

The only hyperparameter for Linear Regression in this experiment
is fit_intercept which can be either True or False.

fit_intercept=True means the model adds a constant term (intercept)
to the equation. This allows the regression line to shift up or down
to better fit the data. This is the standard setting.

fit_intercept=False forces the regression line through the origin,
meaning when all features are 0 the predicted price is also 0. For
car pricing this makes no practical sense - a car does not cost $0
just because all its features happen to be 0 after scaling.

I expect fit_intercept=True to perform better for this reason.
Testing both gives us confirmation of this expectation and shows
the impact of this hyperparameter on model performance.

"""
```

```python
# DO NOT MODIFY THE CODE IN THIS CELL
print_tile(size="h3", key='hyperparameters_selection_explanations', value=hyperparameters_selection_explanations)
```

### C.3 Fit Model

```python
# model with intercept
model_true.fit(X_train, y_train)
# model without intercept
model_false.fit(X_train, y_train)

print("Both models trained successfully")
```

## D. Model Evaluation

### D.1 Model Technical Performance

```python
# evaluate both models
train_rmse_true = rmse(y_train, model_true.predict(X_train))
val_rmse_true = rmse(y_val, model_true.predict(X_val))

train_rmse_false = rmse(y_train, model_false.predict(X_train))
val_rmse_false = rmse(y_val, model_false.predict(X_val))

print(f"With Intercept    — Training RMSE: ${train_rmse_true:,.2f} | Validation RMSE: ${val_rmse_true:,.2f}")
print(f"Without Intercept — Training RMSE: ${train_rmse_false:,.2f} | Validation RMSE: ${val_rmse_false:,.2f}")
```

```python
results_df = pd.DataFrame([
    {"fit_intercept": "Baseline", "train_rmse": 26155.62, "val_rmse": 37291.73},
    {"fit_intercept": "fit_intercept = True", "train_rmse": round(train_rmse_true, 2), "val_rmse": round(val_rmse_true, 2)},
    {"fit_intercept": "fit_intercept = False", "train_rmse": round(train_rmse_false, 2), "val_rmse": round(val_rmse_false, 2)}
])

order = ["Baseline", "With Intercept", "Without Intercept"]

alt.Chart(results_df.melt("fit_intercept", var_name="split", value_name="rmse")).mark_bar().encode(
    alt.X("fit_intercept:N", title="", sort=order, axis=alt.Axis(labelAngle=-15)),
    alt.Y("rmse:Q", title="RMSE ($)"),
    alt.Color("split:N", legend=alt.Legend(title="Split")),
    alt.Column("split:N", sort=order, title=""),
    tooltip=["fit_intercept", "split", "rmse"]
).properties(
    title=alt.TitleParams("Linear Regression RMSE vs Baseline", anchor="middle"),
    width=200,
    height=300
).interactive()
```

```python
coef_df = pd.DataFrame({
    "feature": X_train.columns,
    "coefficient": model_true.coef_.flatten()
}).sort_values("coefficient", ascending=False)

print(f"Intercept: ${model_true.intercept_[0]:,.2f}")
print()
print(coef_df.to_string(index=False))
```

```python
price_ranges = [("<$15k",0,15000),("$15k-$30k",15000,30000),("$30k-$50k",30000,50000),("$50k-$100k",50000,100000),(">$100k",100000,9e9)]

y_actual = y_val.values.flatten()
y_pred = model_true.predict(X_val).flatten()

print(f"{'Segment':<12} {'Count':>6}  {'RMSE':>12}")
print("-" * 35)
for label, lo, hi in price_ranges:
    mask = (y_actual >= lo) & (y_actual < hi)
    if mask.sum() == 0: continue
    rmse = np.sqrt(((y_actual[mask] - y_pred[mask])**2).mean())
    print(f"{label:<12} {mask.sum():>6}  ${rmse:>11,.2f}")
```

```python
model_performance_explanations = """

Two Linear Regression models were trained with different values of fit_intercept and compared
against the baseline RMSE of $37,291.73.

Results:
Baseline          — Train RMSE: $26,155.62 | Val RMSE: $37,291.73
With Intercept    — Train RMSE: $20,071.20 | Val RMSE: $27,162.40
Without Intercept — Train RMSE: $29,715.99 | Val RMSE: $40,393.94

fit_intercept=True clearly outperforms the baseline on both training and validation. The validation
RMSE of $27,162.40 is a $10,129.33 improvement over the baseline.

Performance by price range (with intercept):
Segment       Count          RMSE
-----------------------------------
<$15k             4  $  11,544.82
$15k-$30k       892  $   6,278.85 <- more reliable range
$30k-$50k      1131  $  11,633.54
$50k-$100k      498  $  25,639.13
>$100k           83  $ 130,247.06

The model is only really reliable in the $15k-$30k range. On 50k and above, the error is too high to
be dependable.


Looking at the coefficients:
- brand_bodytype_encoded has the strongest signal at +$13040.47
- kilometres_driven has the strongest negative signal at -$6,878.75
- manufacturing_year has the weakest at +$615 year

"""
```

```python
# DO NOT MODIFY THE CODE IN THIS CELL
print_tile(size="h3", key='model_performance_explanations', value=model_performance_explanations)
```
