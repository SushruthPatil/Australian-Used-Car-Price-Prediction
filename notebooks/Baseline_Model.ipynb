### A.2 Selection of Performance Metrics

> Provide some explanations on why you believe the performance metrics you chose is appropriate


```python
performance_metrics_explanations = """

The performance metric I am using is RMSE (Root Mean Squared Error)

Formula:
RMSE = sqrt( (1/n) * sum( (actual - predicted)^2 ) )
where n is the number of records, actual is the real price and predicted is what the model predicts.

RMSE squares each individual error, takes its average and then square roots it to bring it back to
the original unit (AUD). The squaring step penalises large errors much more than small ones.

This is the right metric for this problem because an RMSE of $5,000 means the model is off by $5,000
which is directly interpretable and meaningful to a non-technical user.

Large errors matter more than small ones in this context as a $20,000 mispricing could cause a buyer
to seriously overpay or walk away from a good deal.

"""
```

```python
# DO NOT MODIFY THE CODE IN THIS CELL
print_tile(size="h3", key='performance_metrics_explanations', value=performance_metrics_explanations)
```

### A.3 Baseline Model Performance

> Provide some explanations on model performance


```python
base_reg = DummyRegressor(strategy="mean")
base_reg.fit(X_train, y_train)

print(f"Mean training price (baseline prediction): ${y_train.mean()[0]:,.2f}")
```

```python
y_train_pred = base_reg.predict(X_train)
y_val_pred = base_reg.predict(X_val)

train_rmse = rmse(y_train, y_train_pred)
val_rmse = rmse(y_val, y_val_pred)

print(f"Training RMSE  : ${train_rmse:,.2f}")
print(f"Validation RMSE: ${val_rmse:,.2f}")
```

```python
baseline_performance_explanations = """

The baseline model, the dummy regressor replaces price with the mean of training
data price, $24,734.24 for every single car in the dataset. It completely ignores brand, km,
age and everything else. The whole point of building a dummy model is to give us a reference point
that our actual models need to beat.

Training RMSE  : $26,155.62
Validation RMSE: $37,291.73

The training RMSE of $26,155.62 means on average the baseline is off by about $26,155.62 per car
which clearly exceeds the threshold of $7,420.27 but it is to be expected from a dummy model.
The validation RMSE jumps to $37,291.73 because the validation set has newer 2019-2020 vehicles that
are generally more expensive than the $24,734.24 training mean. The baseline consistently underestimates
these newer cars since it always predicts the same value.

This is the worst a model can perform. Any experiment that cannot beat $37,291.73 on validation has
basically learned nothing from the data.

"""
```
